# Matrix

```python
import numpy as np
```

```python
# Basic matrix
A = np.array([[1, 2, 3],
             [4, 5, 6]])

print(A.shape) # (2, 3)
print(A.ndim)  # 2
print(A[0, 1]) # 2 <- row 0, col 1
```

```python
# Identity matrix
I = np.eye(3)
print(I)
```

```python
# Zero matrix
Z = np.zeros((3, 4))
print(Z.shape)
print(Z)
```

```python
# Ones matrix
O = np.ones((2, 3))
print(O)
```

```python
# Diagonal matrix
D = np.diag([5, 3, 7])
print(D)
```

#### Dataset as matrix

```python
# Mercari dataset simulation
# 5 products x 4 features
# [prices, conditions, shipping, category]

X = np.array([
    [299, 3, 0, 12],
    [149, 1, 60, 5],
    [999, 4, 0, 8],
    [49, 2, 30, 3],
    [599, 5, 0, 15]
])


print("Shape:", X.shape)
print("Product 1:", X[0]) # row 0
print("All prices:", X[:, 0]) # col 0
```

```python
# Check symmetric matrix
A = np.array([[4, 2, 1],
              [2, 5, 3],
              [1, 3, 6]])

is_symmetric = np.array_equal(A, A.T)
print(is_symmetric) # True
```
