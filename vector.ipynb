```python
import numpy as np
```

```python
# Scalar -> only a number
a = 5
b = 3.14

learning_rate = 0.01

print(f"Scalers: {a, b, learning_rate}")
```

```python
# Create Vectors
v = np.array([3, 5, 4])

# Mercari product vector

product = np.array([299.0, 3, 0, 12])

print(v) # [3, 5, 4]
print(v.shape) # (2, ) <- 2D vector
print(product.shape) # (4, ) <- 4D vector
```

```python
# Vector Properties
v = np.array([3, 5])


print(v.ndim) # 1 -> 1D array
print(v.shape) # (2, ) -> 2 elements
print(v.dtype) # int64 -> data type
print(len(v))  # 2 -> length
```

```python
##### Column vs Row Vector #####


# Row vector -> shape (1, 3)
row = np.array([1, 2, 3])


# Column vector -> (3, 1)
col = np.array([[1], [2], [3]])


# Transpose: row -> column
v = np.array([1, 2, 3])

print(v.T) # same (in 1D transpose remain same)
print(row.T.shape) # (3, 1) -> become a column
```

```python
# Basic Vector OPS
a = np.array([3, 5])
b = np.array([1, 2])

print(a+b) # [4 7] -> addition
print(a-b) # [2 3] -> subtraction
print(3*a) # [9 15] -> scalar multiply
print(np.dot(a, b)) # 13 -> dot product
```
