```python
# Basic Vector OPS
a = np.array([3, 5])
b = np.array([1, 2])

print(a+b) # [4 7] -> addition
print(a-b) # [2 3] -> subtraction
print(3*a) # [9 15] -> scalar multiply
print(np.dot(a, b)) # 13 -> dot product
```

### Vector Operations

```python
# vector addition & subtraction
a = np.array([3, 5])
b = np.array([1, 2])

print(a+b)
print(a-b)

# Word embedding 
king = np.array([0.8, 0.3, 0.1])
man = np.array([0.7, 0.1, 0.0])
woman = np.array([0.2, 0.9, 0.0])

queen_approx = king - man + woman
print(f"Queen_approx: {queen_approx}")
```

```python
# Scalar Multipication
weights = np.array([0.5, 0.3, 0.8])
gradient = np.array([0.1, 0.4, 0.2])
lr = 0.01

# Gradient descent update step 
new_weight = weights - lr * gradient
print(new_weight)
```

```python
# Dot Product (Important)

a = np.array([3, 4])
b = np.array([2, 1])

# method 1 
print(np.dot(a, b)) # 10

# method 2
print(a @ b) # 10 <- prefer this


# Neural network neuron simulation
inputs = np.array([1.0, 2.0, 3.0])
weights = np.array([0.2, 0.5, 0.1])
bias = 0.3

output = inputs @ weights + bias
print(output)
```

```python
# Magnitude (norm)

v = np.array([3, 4])

# Method 1 - manual
mag = np.sqrt(np.sum(v**2))
print(mag) # 5.0


# Method 2 - numpy linalg
print(np.linalg.norm(v)) # 5.0 <- prefer this
```

```python
# cosine similarity (from scratch)

def cosine_similarity(a, b):
    dot = np.dot(a, b)
    mag_a = np.linalg.norm(a)
    mag_b = np.linalg.norm(b)

    return dot / (mag_a * mag_b)


user = np.array([5, 3, 0, 1])
movieA = np.array([4, 3, 1, 0])
movieB = np.array([1, 0, 5, 4])

print(cosine_similarity(user, movieA))  # 0.96 <- similar!
print(cosine_similarity(user, movieB))  # 0.23 <- not similar
```
